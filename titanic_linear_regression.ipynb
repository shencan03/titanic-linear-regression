## Feature Extracting

```python
df.columns
```

```python
features_raw = ["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare"]
```

**Features that weight NEGATIVELY more for WOMEN**

["Parch", "Pclass, "SibSp"]

```python
df['sex_female'] = (df['Sex'] == 'female').astype(int)
```

```python
df["Parch_X_Female"] = -(df['Parch'] * df['sex_female'])
df["SibSp_X_Female"] = -(df["SibSp"] * df["sex_female"])
df["Pclass_X_Female"] = -(df["Pclass"] * df["sex_female"])
```

```python
df.head()
```

# Fill in Median for Ages, Cast Fare Type to Float and Train Model

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
# df["Fare"] = df["Fare"].astype(float)
df["Fare"] = df["Fare"].round(2)
```

```python
X = df[features_raw + ["sex_female", "Parch_X_Female", "Pclass_X_Female", "SibSp_X_Female"]]
```

**Drop Sex Column Since We One-Shot Encoded It**

```python
X.drop(columns="Sex", inplace=True)
```

```python
X.head()
```

```python
y = df["Survived"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
test = pd.read_csv("./data/test.csv")
```

```python
test_original = test
```

```python
test["sex_female"] = (test["Sex"] == "female").astype(int)

test["Parch_X_Female"] = -(test['Parch'] * test['sex_female'])
test["SibSp_X_Female"] = -(test["SibSp"] * test["sex_female"])
test["Pclass_X_Female"] = -(test["Pclass"] * test["sex_female"])
```

```python
test = test[X.columns].copy()
test['Age'] = test['Age'].fillna(test['Age'].median())
```

```python
test.head()
```

```python
test.info()
```

```python
test['Fare'] = test['Fare'].fillna(test['Fare'].median()).round(2)
```

```python
prediction = model.predict(test)
```
